```python
def age_grp(age):
  if age >=0 and age < 10:
        return 1
  elif age < 20:
        return 2
  elif age < 30:
        return 3
  elif age < 40:
        return 4
  elif age < 50:
        return 5
  elif age < 80:
        return 6
  else:
        return 7
```

```python
target = np.zeros((sizes,2) , dtype = "float32")
features = np.zeros((sizes , sample.shape[0] , sample.shape[1] , 1) , dtype = "float32")

for i in range(sizes):
  target[i,0] = age_grp(int(ages[i]))/7
  target[i ,1] = int(genders[i])
  features[i] = imagess[i]
  features[i] = features[i] /255


display(features[1])
print(features[0])
print(target[1,0])
```

now we make the model

```python
import keras
from keras.layers import *
from keras.models import *
from keras import backend as k
```

```python
inputs = Input(shape=(64,64,1))
conv1 = Conv2D(32 ,kernel_size = (3,3),activation="relu")(inputs)
conv2 = Conv2D(64 , kernel_size = (3,3) , activation = "relu")(conv1)
pool1 = MaxPooling2D(pool_size=(2,2))(conv2)
conv3 = Conv2D(128 , kernel_size = (3,3) , activation = "relu")(pool1)
pool2 = AveragePooling2D(pool_size = (2,2))(conv3)
conv4 = Conv2D(256,kernel_size = (3,3),activation="relu")(pool2)
pool3 = MaxPooling2D(pool_size = (2,2))(conv4)
x = Dropout(0.25)(pool3)
flat = Flatten()(x)

# fc for age
dropout = Dropout(0.5)
age_model = Dense(256, activation='relu')(flat)
age_model = dropout(age_model)
age_model = Dense(128 , activation = "relu")(flat)
age_model = dropout(age_model)
age_model = Dense(64 , activation = "relu")(flat)
age_model = dropout(age_model)
age_model = Dense(32 , activation = "relu")(flat)
age_model = dropout(age_model)
age_model = Dense(1 , activation = "relu")(flat)


#fc for gender
dropout = Dropout(0.5)
gender_model = Dense(256, activation='relu')(flat)
gender_model = dropout(gender_model)
gender_model = Dense(128 , activation = "relu")(flat)
gender_model = dropout(gender_model)
gender_model = Dense(64 , activation = "relu")(flat)
gender_model = dropout(gender_model)
gender_model = Dense(32 , activation = "relu")(flat)
gender_model = dropout(gender_model)
gender_model = Dense(1 , activation = "sigmoid")(flat)

```

```python
model = Model(inputs=inputs , outputs=[age_model ,gender_model])
model.compile(optimizer = "adam" , loss = ["mse" , "binary_crossentropy"],metrics=["accuracy"])
```
